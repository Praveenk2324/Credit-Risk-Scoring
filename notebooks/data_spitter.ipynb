```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv(r"..\data\raw\GiveMeSomeCredit-training.csv")

os.makedirs('../data/simulation_stream', exist_ok=True)

base_data = df.iloc[:100000].copy()
base_data.to_csv('../data/simulation_stream/01_base_data.csv')

batch_1 = df.iloc[100000:125000].copy()
batch_1.to_csv('../data/simulation_stream/02_batch_1_normal.csv')

batch_2 = df.iloc[125000:].copy()
batch_2['age'] = batch_2['age'] - 15  # Shift age way down
batch_2['age'] = batch_2['age'].apply(lambda x: max(18, x)) # Keep it legal
batch_2['DebtRatio'] = batch_2['DebtRatio'] * 5 # Explode the debt ratio

batch_2.to_csv('../data/simulation_stream/03_batch_2_drifted.csv')

print("Simulation stream created successfully!")
```
